```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sn
import time

%matplotlib notebook

sn.set_style("ticks",sn.axes_style({'axes.grid': True}))

def distance(a,b):
    d = np.sqrt( (b[0]-a[0])**2 + (b[1]-a[1])**2 + (b[2]-a[2])**2 )
    return d
```

```python
data = np.loadtxt('500k_weighted_north_cmass.dat')

dt=data.transpose()
```

```python
from spherical_to_cartesian import spherical_to_cartesian as s2c

import astropy
#from astropy.cosmology import WMAP9 as cosmo
from astropy.cosmology import FlatLambdaCDM

cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
cd = cosmo.comoving_distance(dt[2])
cdist = cd.value * 0.7

d_conv=s2c(dt,cdist)
```

```python
plt.figure()
sn.kdeplot(d_conv[1],d_conv[2],bw='silverman',shade=True,cmap='inferno',n_levels=125)
plt.savefig('sdss_density.png')
```
