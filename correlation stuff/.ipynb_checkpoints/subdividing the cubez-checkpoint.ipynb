```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import time
%matplotlib notebook
sn.set_style("ticks",sn.axes_style({'axes.grid': True}))
```

```python
random = np.loadtxt('Gravitationally_Rigid_Intergalactic_Dimensions_10k_random.dat')
data = np.loadtxt('Gravitationally_Rigid_Intergalactic_Dimensions_10k_data.dat')
```

```python
dt = data.transpose()
```

```python
div1=[]
div2=[]
div3=[]
div4=[]
div5=[]
div6=[]
div7=[]
div8=[]

mx = np.median((min(dt[0]),max(dt[0])))
my = np.median((min(dt[1]),max(dt[1])))
mz = np.median((min(dt[2]),max(dt[2])))


for i in data:
    
    if   i[0]<=mx and i[1]<=my and i[2]<=mz:
        div1.append(i)
    elif i[0]<=mx and i[1]<=my and i[2]> mz:
        div2.append(i)
    elif i[0]<=mx and i[1]> my and i[2]<=mz:
        div3.append(i) 
    elif i[0]> mx and i[1]<=my and i[2]<=mz:
        div4.append(i)
    elif i[0]> mx and i[1]> my and i[2]<=mz:
        div5.append(i)
    elif i[0]> mx and i[1]<=my and i[2]> mz:
        div6.append(i)
    elif i[0]<=mx and i[1]> my and i[2]> mz:
        div7.append(i)
    elif i[0]> mx and i[1]> my and i[2]> mz:
        div8.append(i)
        
print(len(div1)+len(div2)+len(div3)+len(div4)+len(div5)+len(div6)+len(div7)+len(div8))
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(np.array(div1).transpose()[0],np.array(div1).transpose()[1],np.array(div1).transpose()[2],s=30,color='k')
ax.scatter(np.array(div2).transpose()[0],np.array(div2).transpose()[1],np.array(div2).transpose()[2],s=30)
ax.scatter(np.array(div3).transpose()[0],np.array(div3).transpose()[1],np.array(div3).transpose()[2],s=30)
ax.scatter(np.array(div4).transpose()[0],np.array(div4).transpose()[1],np.array(div4).transpose()[2],s=30)
ax.scatter(np.array(div5).transpose()[0],np.array(div5).transpose()[1],np.array(div5).transpose()[2],s=30)
ax.scatter(np.array(div6).transpose()[0],np.array(div6).transpose()[1],np.array(div6).transpose()[2],s=30)
ax.scatter(np.array(div7).transpose()[0],np.array(div7).transpose()[1],np.array(div7).transpose()[2],s=30)
ax.scatter(np.array(div8).transpose()[0],np.array(div8).transpose()[1],np.array(div8).transpose()[2],s=30)
```

```python
#dd

def distance(a,b):
    dist = np.sqrt( (b[0]-a[0])**2 + (b[1]-a[1])**2 + (b[2]-a[2])**2 )
    return dist

dd=[]

start = time.time()

###############################################################################

alldivs = [div1,div2,div3,div4,div5,div6,div7,div8]

### subdiv-subdiv ###

for idiv in range(0,len(alldivs)-1):
    
    subdiv = alldivs[idiv]
    
    for jdiv in range(idiv+1,8):
    
        subdiv2 = alldivs[jdiv]
        
        for j in range(0,len(subdiv)):
        
            pos1=subdiv[j]
            pos2=np.array(subdiv2[:]).transpose()
            dist=distance(pos1,pos2)
            dd += dist.tolist()
            
### subdiv-self ###
    
for idiv in range(0,len(alldivs)):
    
    subdiv = alldivs[idiv]
    
    for i in range(0,len(subdiv)-1):
        pos1=subdiv[i]
        pos2=np.array(subdiv[i+1:]).transpose()
        dist=distance(pos1,pos2)
        dd+=dist.tolist()

###############################################################################

step1=time.time()

print("Done! Took %.1f seconds." % (step1-start))

print (len(dd))

plt.figure()
hist=plt.hist(dd,bins=750)
sn.despine()

###############################################################################

### FOR COMPARISON ###

start = time.time()

dd2=[]

for i in range(0,len(data)-1):
    pos1 = data[i]
    pos2 = data[i+1:].transpose()
    dist = distance(pos1,pos2)
    dd2 += dist.tolist()
        
step1=time.time()
     
print("Done! (DD) Took %.1f seconds." %(step1-start))

plt.figure()
hist=plt.hist(dd2,bins=750)
sn.despine()
print(len(dd2))

###############################################################################

```
