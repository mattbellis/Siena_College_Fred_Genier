```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sn
import time

%matplotlib notebook

sn.set_style("ticks",sn.axes_style({'axes.grid': True}))

def distance(a,b):
    d = np.sqrt( (b[0]-a[0])**2 + (b[1]-a[1])**2 + (b[2]-a[2])**2 )
    return d

g1 = (210.88543148, -1069.32087983,   470.3136259)
g2 = (224.51380802, -1020.26707054,   490.96604484)
obs = (0,0,0)
```

```python
ax = plt.axes(projection="3d")
ax.scatter(obs[0],obs[1],obs[2])
ax.scatter(g1[0],g1[1],g1[2])
ax.scatter(g2[0],g2[1],g2[2])
```

```python
d1 = distance(obs,g1)
d2 = distance(obs,g2)
d3 = distance(g1 ,g2)

print(d1,d2,d3)

#s is the shortest side, m is the middle side
s = min(d1,d2,d3)
m = d1
```
