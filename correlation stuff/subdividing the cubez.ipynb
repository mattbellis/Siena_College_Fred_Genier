```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import time
%matplotlib notebook
sn.set_style("ticks",sn.axes_style({'axes.grid': True}))
```

```python
random = np.loadtxt('Gravitationally_Rigid_Intergalactic_Dimensions_10k_random.dat')
data = np.loadtxt('Gravitationally_Rigid_Intergalactic_Dimensions_10k_data.dat')
```

```python
dt = data.transpose()
rt = random.transpose()
```

```python
### Data ###

div1=[]
div2=[]
div3=[]
div4=[]
div5=[]
div6=[]
div7=[]
div8=[]

mx = np.median((min(dt[0]),max(dt[0])))
my = np.median((min(dt[1]),max(dt[1])))
mz = np.median((min(dt[2]),max(dt[2])))


for i in data:
    
    if   i[0]<=mx and i[1]<=my and i[2]<=mz:
        div1.append(i)
    elif i[0]<=mx and i[1]<=my and i[2]> mz:
        div2.append(i)
    elif i[0]<=mx and i[1]> my and i[2]<=mz:
        div3.append(i) 
    elif i[0]> mx and i[1]<=my and i[2]<=mz:
        div4.append(i)
    elif i[0]> mx and i[1]> my and i[2]<=mz:
        div5.append(i)
    elif i[0]> mx and i[1]<=my and i[2]> mz:
        div6.append(i)
    elif i[0]<=mx and i[1]> my and i[2]> mz:
        div7.append(i)
    elif i[0]> mx and i[1]> my and i[2]> mz:
        div8.append(i)
        
print(len(div1)+len(div2)+len(div3)+len(div4)+len(div5)+len(div6)+len(div7)+len(div8))
```

```python
### Random ###

rdiv1=[]
rdiv2=[]
rdiv3=[]
rdiv4=[]
rdiv5=[]
rdiv6=[]
rdiv7=[]
rdiv8=[]

mx = np.median((min(rt[0]),max(rt[0])))
my = np.median((min(rt[1]),max(rt[1])))
mz = np.median((min(rt[2]),max(rt[2])))


for i in random:
    
    if   i[0]<=mx and i[1]<=my and i[2]<=mz:
        rdiv1.append(i)
    elif i[0]<=mx and i[1]<=my and i[2]> mz:
        rdiv2.append(i)
    elif i[0]<=mx and i[1]> my and i[2]<=mz:
        rdiv3.append(i) 
    elif i[0]> mx and i[1]<=my and i[2]<=mz:
        rdiv4.append(i)
    elif i[0]> mx and i[1]> my and i[2]<=mz:
        rdiv5.append(i)
    elif i[0]> mx and i[1]<=my and i[2]> mz:
        rdiv6.append(i)
    elif i[0]<=mx and i[1]> my and i[2]> mz:
        rdiv7.append(i)
    elif i[0]> mx and i[1]> my and i[2]> mz:
        rdiv8.append(i)
        
print(len(rdiv1)+len(rdiv2)+len(rdiv3)+len(rdiv4)+len(rdiv5)+len(rdiv6)+len(rdiv7)+len(rdiv8))
```

```python
def distance(a,b):
    dist = np.sqrt( (b[0]-a[0])**2 + (b[1]-a[1])**2 + (b[2]-a[2])**2 )
    return dist

ddivs=[div1,div2,div3,div4,div5,div6,div7,div8]
rdivs=[rdiv1,rdiv2,rdiv3,rdiv4,rdiv5,rdiv6,rdiv7,rdiv8]
```

```python
### DD ###

dd=[]

start = time.time()

### subdiv-subdiv ###

for idiv in range(0,len(ddivs)-1):
    
    subdiv = ddivs[idiv]
    
    for jdiv in range(idiv+1,8):
    
        subdiv2 = ddivs[jdiv]
        
        for j in range(0,len(subdiv)):
        
            pos1=subdiv[j]
            pos2=np.array(subdiv2[:]).transpose()
            dist=distance(pos1,pos2)
            dd += dist.tolist()
            
### subdiv-self ###
    
for idiv in range(0,len(ddivs)):
    
    subdiv = ddivs[idiv]
    
    for i in range(0,len(subdiv)-1):
        pos1=subdiv[i]
        pos2=np.array(subdiv[i+1:]).transpose()
        dist=distance(pos1,pos2)
        dd+=dist.tolist()

step1=time.time()

print("Done (DD)! Took %.1f seconds." % (step1-start))

print ("%.2e datapoints." % (len(dd)))

plt.figure()
hist_dd=plt.hist(dd,bins=200)
sn.despine()
print(len(dd))
```

```python
### DR ###
dr=[]

start = time.time()

for idiv in range(0,8):
    
    subdiv1 = ddivs[idiv]
    
    for jdiv in range(0,8):
        
        #print(idiv+1,jdiv+1) #debug
    
        subdiv2 = rdivs[jdiv]
        
        for j in range(0,len(subdiv1)):
        
            pos1=subdiv1[j]
            pos2=np.array(subdiv2).transpose()
            dist=distance(pos1,pos2)
            dr += dist.tolist()
            
end = time.time()

print("Done! (DR) Took %.2f seconds"%(end-start))
print("%.2e datapoints."%(len(dr)))

plt.figure()
hist_dr=plt.hist(dr,bins=200)
sn.despine()
print(len(dr))
```

```python
### RR ###

rr=[]

start = time.time()

### subdiv-subdiv ###

for idiv in range(0,len(rdivs)-1):
    
    subdiv = rdivs[idiv]
    
    for jdiv in range(idiv+1,8):
    
        subdiv2 = rdivs[jdiv]
        
        for j in range(0,len(subdiv)):
        
            pos1=subdiv[j]
            pos2=np.array(subdiv2[:]).transpose()
            dist=distance(pos1,pos2)
            rr += dist.tolist()
            
### subdiv-self ###
    
for idiv in range(0,len(rdivs)):
    
    subdiv = rdivs[idiv]
    
    for i in range(0,len(subdiv)-1):
        pos1=subdiv[i]
        pos2=np.array(subdiv[i+1:]).transpose()
        dist=distance(pos1,pos2)
        rr+=dist.tolist()


end = time.time()

print("Done (RR)! Took %.1f seconds." % (end-start))

print ("%.2e datapoints." % (len(rr)))

plt.figure()
hist_rr=plt.hist(rr,bins=200)
sn.despine()
print(len(rr))
```
